```python
import os
import shutil
import yaml
from google.colab import drive
from ultralytics import YOLO

# Monta o Drive
drive.mount('/content/drive')

NOME_ZIP_NO_DRIVE = "License Plate Recognition.v4-resized640_aug3x-accurate.yolov11.zip"
CAMINHO_ZIP_DRIVE = f"/content/drive/MyDrive/License Plate Recognition.v4-resized640_aug3x-accurate.yolov11/{NOME_ZIP_NO_DRIVE}"

PASTA_LOCAL = "/content/dataset_placas"

if not os.path.exists(PASTA_LOCAL):
    print(f"Copiando {NOME_ZIP_NO_DRIVE} do Drive para o Colab...")
    shutil.copy(CAMINHO_ZIP_DRIVE, "/content/temp.zip")

    print("Descompactando...")
    shutil.unpack_archive("/content/temp.zip", PASTA_LOCAL)

    os.remove("/content/temp.zip") # Limpa lixo
    print("Dataset pronto para uso!")
else:
    print("Dataset jÃ¡ estava descompactado.")

arquivo_yaml = f"{PASTA_LOCAL}/data.yaml"

if not os.path.exists(arquivo_yaml):
    for root, dirs, files in os.walk(PASTA_LOCAL):
        if "data.yaml" in files:
            arquivo_yaml = os.path.join(root, "data.yaml")
            break

print(f"Usando arquivo de configuraÃ§Ã£o: {arquivo_yaml}")

with open(arquivo_yaml, 'r') as f:
    data_config = yaml.safe_load(f)

pasta_raiz_dataset = os.path.dirname(arquivo_yaml)
data_config['train'] = os.path.join(pasta_raiz_dataset, "train/images")
data_config['val']   = os.path.join(pasta_raiz_dataset, "valid/images")
if 'test' in data_config:
    data_config['test']  = os.path.join(pasta_raiz_dataset, "test/images")

with open(arquivo_yaml, 'w') as f:
    yaml.dump(data_config, f)

model = YOLO("yolo11n.pt")

results = model.train(
    data=arquivo_yaml,  # Caminho do arquivo
    epochs=20,          # Quantidade Ã‰pocas
    patience=3,         # Early stopping
    imgsz=640,          # Tamanho das imagens
    batch=32,           # Quantidade de imagens que irÃ¡ analisar atÃ© ajustar os pesos do modelo
    device=0,           # 0 = GPU
    workers=8,          # Quantidade de nÃºcleos do processador
    cache=False,        # NÃ£o guardar cache pela quantidade de imagens
    project="/content/drive/MyDrive/Plate Recognition",
    name="modelo_identificador_placas",
    exist_ok=True       # Utiliza as pastas jÃ¡ criadas se elas jÃ¡ existirem
)
```
